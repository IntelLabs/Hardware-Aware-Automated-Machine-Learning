# 32 Threads running on 8 cores, AVX

```python
RESULTS_FILE = 'experiment_results/2024-09-11_01-14-31/results.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(RESULTS_FILE)

# Filter the data for context length of 512 and generated tokens of 128
filtered_df = df[(df['Context Length'] == 512) & (df['Number of generated Tokens'] == 128)]

# Group the filtered data, calculate mean Decode Latency, and convert to milliseconds
grouped_df = filtered_df.groupby('Number of Column Groups')['Decode Latency'].mean().reset_index()
grouped_df['Decode Latency'] *= 1000  # Convert seconds to milliseconds

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=grouped_df, x='Number of Column Groups', y='Decode Latency', marker='o')

# Customize the plot
plt.title('Decode Latency vs Number of Column Groups\n(Context Length: 512, Generated Tokens: 128) - 32 Threads - 8 Cores')
plt.xlabel('Number of Column Groups')
plt.ylabel('Decode Latency (ms)')
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

# Comprehensive Results - AVX & AMX

```python
RESULTS_FILE = 'experiment_results/2024-09-11_16-26-31/results.csv'
# Load the CSV file
df = pd.read_csv(RESULTS_FILE)

# Display the first few rows of the dataframe
df.head()
```

```python
# Filter the data for 8 threads and cores 0-7
filtered_df = df[(df['Number of Threads'] == 8) & (df['Number of Cores'] == '0-7')]

# Group the filtered data, calculate mean Decode Latency, and convert to milliseconds
grouped_df = filtered_df.groupby(['Number of Column Groups', 'Mode'])['Decode Latency'].mean().reset_index()
grouped_df['Decode Latency'] *= 1000  # Convert seconds to milliseconds

# Create the plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped_df, x='Number of Column Groups', y='Decode Latency', hue='Mode', marker='o')

# Customize the plot
plt.title('Decode Latency vs Number of Column Groups\n(8 Threads, Cores 0-7)')
plt.xlabel('Number of Column Groups')
plt.ylabel('Decode Latency (ms)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Mode')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter the data for 8 threads and cores 0-7
filtered_df = df[(df['Number of Threads'] == 8) & (df['Number of Cores'] == '0-7')]

# Group the filtered data, calculate mean Decode Latency
grouped_df = filtered_df.groupby(['Number of Column Groups', 'Mode'])['Decode Latency'].mean().reset_index()

# Calculate tokens per second
grouped_df['Tokens per Second'] = 1 / grouped_df['Decode Latency']

# Create two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

# Plot 1: Decode Latency
sns.lineplot(data=grouped_df, x='Number of Column Groups', y='Decode Latency', hue='Mode', marker='o', ax=ax1)
ax1.set_title('Decode Latency vs Number of Column Groups\n(8 Threads, Cores 0-7)')
ax1.set_xlabel('Number of Column Groups')
ax1.set_ylabel('Decode Latency (s)')
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(title='Mode')

# Plot 2: Tokens per Second
sns.lineplot(data=grouped_df, x='Number of Column Groups', y='Tokens per Second', hue='Mode', marker='o', ax=ax2)
ax2.set_title('Tokens per Second vs Number of Column Groups\n(8 Threads, Cores 0-7)')
ax2.set_xlabel('Number of Column Groups')
ax2.set_ylabel('Tokens per Second')
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend(title='Mode')

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
# Group by Cores and Mode, find the minimum Decode Latency
best_performance = df.groupby(['Number of Cores', 'Mode'])['Decode Latency'].min().reset_index()

# Function to extract the maximum core number
def max_core(core_range):
    return max(int(x) for x in core_range.replace(' ', '').split('-'))

# Apply the function to create a new column for sorting
best_performance['Max Core'] = best_performance['Number of Cores'].apply(max_core)

# Sort the dataframe
best_performance_sorted = best_performance.sort_values('Max Core')
best_performance_sorted['Decode Latency'] *= 1000  # Convert seconds to milliseconds


# Create the plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=best_performance_sorted, x='Number of Cores', y='Decode Latency', hue='Mode', marker='o')

# Customize the plot
plt.title('Best Decode Latency for Each Core Configuration by Mode')
plt.xlabel('Cores')
plt.ylabel('Decode Latency (ms)')
plt.yscale('log')  # Using log scale for better visualization of differences
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Mode')


plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.0f}'))


# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout and show plot
plt.tight_layout()
plt.show()

# Display the data
print(best_performance_sorted)
```
